#### Performing Calculations on ints and floats

##### Addition

```python
num1 = 2
num2 = 5.4

result = num1 + num2

print(result)

result += 2
print(result)
```

##### Subtraction

```python
result_diff = num2 - num1

result_diff -= 1
print(result_diff - 1)
```

##### Multiplication

```python
result_mul = num1 * num2
print(result_mul)

result_mul *= 2
print(result_mul)
```

##### Division

```python
result_div = num2/num1
print(result_div)

result_div /= 3
print(result_div)
```

##### Floor Division

```python
result_floor = num2 // num1 
print(result_floor)

result_floor //= 2
print(result_floor)
```

##### Modulo

```python
result_mod = 13 % 2
print(result_mod)

result_mod %=2
print(result_mod)
```

##### Exponential

```python
square = 5 ** 2
print(square)

square **= 2
print(square)
```

### If Statements

```python
# Available operators: Greater(>), Less(<),Equal(==)
# Greater or Equal(>=), Less or Equal (<=)

# Truth Tree:
# T && F = F
# T && T = T
# T || F = T
# F || T = T
# F || F = F

num1 = 5
num2 = 10
num3 = 0
# if num1 == num2:
#     print('Equal Values')
# else:
#     print('Not Equal')
if num2 > num1 or num3 > 0:
    print('This ran!')
elif num1 > num2:
    print('Num1 is greater')
else:
    print('You are here')
```

## Using 'and'/'or' with If Statements

```python
num_11 = 15
num_12 = 3
num_13 = 10
num_14 = 3

# If with 'and' statement
if num_11 / 5 ==  num_12 and num_13 - 7 == num_14:
    print("True and True")

# If with 'or' statement
if num_11 > num_12 or num_13 == num_14:
    print('True and False')
```

### While Loops

```python
# Syntax:
# While keyword, condition statement

num = 0

while num < 10:
    print(num)
    num += 1
```

##### Looping 'While True'

```python
game_over = True

while game_over:
    print('Infinite Loop')
    user_input = input('Would you like to stop? ')
    if user_input == 'Yes':
        game_over = False
```

##### While & For Loops Used Together

```python
num = 0

while num < 5:
    print("\nWhile loop iteration: " + str(num))
    
    for i in range(2):
        print("For loop iteration: ", str(i))
    
    num += 1
```

### Lists

##### Declaring Lists

```python
list_1 = []

names = ['Max', 'Cindy', 'Kathy', "bob", 'Nate']
print(names)
```

##### Indexing a List

```python
# list_name[start: stop: step]

# Single Index
print(names[1])

# Print starting at index 1 going to the end
print(names[1:])

# Print starting at the beginning of a list up until a number
print(names[:2])

# Print starting at index 1 and going up by 2 in each iteration
print(names[1::2])

# Print starting at the back and display in reverse order
print(names[::-1])
```

##### .append()

```python
names.append('Josias')
print(names)
```

##### .insert()

```python
names.insert(3, 'Devon')
print(names)
```

##### .pop()

```python
# Defaults to the last value if no parameter is given
# Pop returns the element that was removed in casee you want to assign it to a variable

my_name = names.pop(2)
print(my_name)
print(names)
```


##### .remove()

```python
# Value to be removed rather than index
# names.remove('bob')
# print(names)

while 'Josias' in names:
    names.remove('Josias')
print(names)
```

##### del()

```python
# Goes by index, rather than value
# Be careful with del, it can cause indexing errors if not used carefully

del(names[1])
print(names)
```

##### Concatenating Two Lists

```python
# Will append two lists together, will NOT add the values!!

list_2 = [0, 1, 2]
list_3 = [3, 4, 5]

large_list = list_2 + list_3
print(large_list)
```

##### Lists Within Lists

```python
# List can hold ANY type of other element! Including other lists!
# They can go as deep as you want; this is called nested lists

names = ["Max", "sam", "Josh", ["Sally", ["Sue", "Jim"], "Tameka"]]
print(names)
print(names[3][1][1])
```

##### Looping Through Lists

```python
# Two ways to loop through a list! One is by index; the other is by using the 'in' keyword

# By index
for i in range(len(names)):
    print(names[i])
    
# Loop with 'in'
for i in names:
    print(i)
```

## Exercise #2 <br>
<p>Get first prime numbers up to 100</p>

```python
# HINT::
# An else after an if runs if the if didn’t
# An else after a for runs if the for didn’t break

for a in range(1, 101):
    me = True
    for b in range(2, a):
        if a % b == 0:
            me = False
        if me == True:
            print(a)
            break
```

# Exercise 3 <br>
<p>Take in a users input for their age, if they are younger than 18 print kids, if they're 18 to 65 print adults, else print seniors</p>

```python
age = int(input("What is your age? "))

# If a person is younger than 18
if age <= 18:
    print('Kids')

# If a person is 18 to 65
if age >= 18 and age <= 65:
    print('Adults')
else:
  if age > 65:
    print('Seniors')
```
